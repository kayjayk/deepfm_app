```python
genre_col = ['meta_genre', 'genre_large', 'genre_mid', 'genre_small']
for genre in tqdm.tqdm(final_genre):
    for col in genre_col:
        idx = np.where(raw[col].str.contains(genre))
        raw[genre].loc[idx] = True
```
