```python
genre_col = 'meta_genre'
genres_list = list(raw[genre_col].apply(lambda x: str(x).strip().split('/')))

flat_list = []
for sub_list in genres_list:
    for genre in sub_list:
        if genre not in flat_list:
            flat_list.append(genre)

for i in range(len(flat_list)):
    flat_list[i] = flat_list[i].split(', ')

flat_list_2 = []
for sub_list in flat_list:
    for genre in sub_list:
        if genre not in flat_list_2:
            flat_list_2.append(genre)

for i in range(len(flat_list_2)):
    flat_list_2[i] = flat_list_2[i].split(',')
    
flat_list_3 = []
for sub_list in flat_list_2:
    for genre in sub_list:
        if genre not in flat_list_3:
            flat_list_3.append(genre)

# raw[genre_col] = raw[genre_col].astype(str)

meta_genre = flat_list_3

genre_col = 'genre_large'
genres_list = list(raw[genre_col].apply(lambda x: str(x).strip().split('/')))

flat_list = []
for sub_list in genres_list:
    for genre in sub_list:
        if genre not in flat_list:
            flat_list.append(genre)

for i in range(len(flat_list)):
    flat_list[i] = flat_list[i].split(', ')

flat_list_2 = []
for sub_list in flat_list:
    for genre in sub_list:
        if genre not in flat_list_2:
            flat_list_2.append(genre)

for i in range(len(flat_list_2)):
    flat_list_2[i] = flat_list_2[i].split(',')
    
flat_list_3 = []
for sub_list in flat_list_2:
    for genre in sub_list:
        if genre not in flat_list_3:
            flat_list_3.append(genre)

# raw[genre_col] = raw[genre_col].astype(str)

large_genre = flat_list_3

genre_col = 'genre_mid'
genres_list = list(raw[genre_col].apply(lambda x: str(x).strip().split('/')))

flat_list = []
for sub_list in genres_list:
    for genre in sub_list:
        if genre not in flat_list:
            flat_list.append(genre)

for i in range(len(flat_list)):
    flat_list[i] = flat_list[i].split(', ')

flat_list_2 = []
for sub_list in flat_list:
    for genre in sub_list:
        if genre not in flat_list_2:
            flat_list_2.append(genre)

for i in range(len(flat_list_2)):
    flat_list_2[i] = flat_list_2[i].split(',')
    
flat_list_3 = []
for sub_list in flat_list_2:
    for genre in sub_list:
        if genre not in flat_list_3:
            flat_list_3.append(genre)

# raw[genre_col] = raw[genre_col].astype(str)

mid_genre = flat_list_3

genre_col = 'genre_small'
genres_list = list(raw[genre_col].apply(lambda x: str(x).strip().split('/')))

flat_list = []
for sub_list in genres_list:
    for genre in sub_list:
        if genre not in flat_list:
            flat_list.append(genre)

for i in range(len(flat_list)):
    flat_list[i] = flat_list[i].split(', ')

flat_list_2 = []
for sub_list in flat_list:
    for genre in sub_list:
        if genre not in flat_list_2:
            flat_list_2.append(genre)

for i in range(len(flat_list_2)):
    flat_list_2[i] = flat_list_2[i].split(',')
    
flat_list_3 = []
for sub_list in flat_list_2:
    for genre in sub_list:
        if genre not in flat_list_3:
            flat_list_3.append(genre)

# raw[genre_col] = raw[genre_col].astype(str)

small_genre = flat_list_3
```

```python
print(meta_genre)
print(large_genre)
print(mid_genre)
print(small_genre)
```

```python
all_genre = set(meta_genre + large_genre + mid_genre + small_genre)
```

```python
final_genre = all_genre.difference(('N', 'n', 'ㅜ',
                  '극장판 애니', '극장판 에니', '극장판애니',
                  '다큐멘터리', '독립영화', '애니메이션'))
```

```python
len(final_genre)
```

```python
final_genre
```

```python
for genre in final_genre:
    raw[genre] = False
```

```python
raw
```

```python
genre_col = ['meta_genre', 'genre_large', 'genre_mid', 'genre_small']
for genre in tqdm.tqdm(final_genre):
    for col in genre_col:
        idx = np.where(raw[col].str.contains(genre))
        raw[genre].loc[idx] = True
```

```python
raw.drop(columns='s_time', inplace=True)
```

* column name str -> int

```python
re_dict = {}
final_genre = list(final_genre)
for i in range(len(final_genre)):
    re_dict[final_genre[i]] = str(i)
```

```python
raw = raw.rename(columns = re_dict)
```
