```python
import pandas as pd
import tqdm
import numpy as np
import pickle
```

```python
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 100)
```

```python
data = pd.read_csv('../../Data/movie_201130_table_3.csv')
```

```python
str_col = ['sa_id', 'album_id']
int_col = ['release_date', 'run_time', 'buy_tot', 'pr_info']
float_col = ['buy_history_price', 'movie_meta_price', 'i30_meta_price']
```

```python
data[str_col] = data[str_col].astype(str)
data[int_col] = data[int_col].astype(int)
data[float_col] = data[float_col].astype(float)
```

```python
data['sa_id'] = data['sa_id'].apply(lambda x: str(x))
data['album_id'] = data['album_id'].apply(lambda x: str(x))
```

```python
data.to_csv('../../Data/movie_201130_table_3_4.csv', index=None)
```
